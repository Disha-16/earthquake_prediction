Ensemble learning

1.Hard Voting

```python
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Example dataset
# Replace x_data and y_data with your actual dataset
# x_data = ...
# y_data = ...

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=44)

# Step 1: Set up individual classifiers
l1 = LogisticRegression()
svm = SVC()
nb = GaussianNB()
dt = DecisionTreeClassifier()

# Step 2: Create the VotingClassifier with hard voting
voting_clf = VotingClassifier(estimators=[
    ("logisticRegression", l1),
    ("svm", svm),
    ("naivebayes", nb),
    ("decisionTree", dt)
], voting='hard')

# Step 3: Fit the VotingClassifier to the training data
voting_clf.fit(X_train, y_train)

# Step 4: Make predictions on the test data
y_pred4 = voting_clf.predict(X_test)

# Step 5: Evaluate performance
accuracy4 = accuracy_score(y_test, y_pred4)
precision = precision_score(y_test, y_pred4, average='binary') # Change to 'macro' or 'micro' if multiclass
recall = recall_score(y_test, y_pred4, average='binary') # Change to 'macro' or 'micro' if multiclass
f1 = f1_score(y_test, y_pred4, average='binary') # Change to 'macro' or 'micro' if multiclass

# Print the results
print("Accuracy:", accuracy4)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

2.soft Voting

```python
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Example dataset
# Replace x_data and y_data with your actual dataset
# x_data = ...
# y_data = ...

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.07424, random_state=38)

# Step 1: Set up individual classifiers
l1 = LogisticRegression()
svm = SVC(probability=True)  # Enable probability estimates for SVM
nb = GaussianNB()
dt = DecisionTreeClassifier()

# Step 2: Create the VotingClassifier with soft voting
voting_clf = VotingClassifier(estimators=[
    ("logisticRegression", l1),
    ("svm", svm),
    ("naivebayes", nb),
    ("decisionTree", dt)
], voting='soft')

# Step 3: Fit the VotingClassifier to the training data
voting_clf.fit(X_train, y_train)

# Step 4: Make predictions on the test data
y_pred5 = voting_clf.predict(X_test)


# Step 5: Evaluate performance
accuracy5 = accuracy_score(y_test, y_pred5)
precision = precision_score(y_test, y_pred5, average='binary')  # Change to 'macro' or 'micro' if multiclass
recall = recall_score(y_test, y_pred5, average='binary')  # Change to 'macro' or 'micro' if multiclass
f1 = f1_score(y_test, y_pred5, average='binary')  # Change to 'macro' or 'micro' if multiclass

# Print the results
print("Accuracy:", accuracy5)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

```python
models = ["Hard Voting","Soft Voting"]
accuracies = [accuracy4,accuracy5]
```
