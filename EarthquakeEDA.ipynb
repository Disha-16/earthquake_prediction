```python
df['date_time'] = pd.to_datetime(df['date_time'], dayfirst=True)
df.info()
```

```python
df['year'] = df['date_time'].dt.year
# extracting month (by name) component from column converted to datetime
df["month"] = df["date_time"].dt.month_name()

df.head(10)
```

```python
col = df.pop('month')
df.insert(3, col.name, col)
col = df.pop('year')
df.insert(3, col.name, col)

df.head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# finding the number of earthquakes per country, to determine the most seismogenic countries:

countries = df['country'].value_counts()
print(countries)
```

```python
top_10_countries = (df['country'].value_counts()).iloc[:10]
print(top_10_countries)
```

```python
top_10_countries.plot(kind='bar')

for index, value in enumerate(top_10_countries):
    plt.text(index, value, str(value), ha='center', va='bottom')

plt.title('Top 10 Countries with Most Earthquakes Since 2001')
plt.xticks(rotation=60, horizontalalignment='right')

plt.show()
```

```python
continents = (df['continent'].value_counts())
print(continents)
```

```python
continents.plot()

for index, value in enumerate(continents):
    plt.text(index, value, str(value), ha='center', va='bottom')
plt.title('Count of Earthquakes by Continent Since 2001')

plt.xticks(rotation=60, horizontalalignment='right')
```

```python
import plotly.express as px

fig = px.scatter_geo(
    df,
    lat='latitude',
    lon='longitude',
    color='magnitude',
    size='depth',
    hover_name='location',
    projection='natural earth',
    title='Earthquake Occurrences Worldwide'
)

fig.show()
```
